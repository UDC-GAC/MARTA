```python
df = []
for f in glob.glob("asm_*.csv"):  
    df.append(pd.read_csv(f))
df = pd.concat(df).reset_index().drop("index",axis=1).drop("NONE",axis=1).rename({"_ASM_PERMUTATION":"N_FMA"},axis=1)
df["arch"] = "intel_cascadelake_silver"
full_df = df
df = []
for f in glob.glob("amd_*.csv"):  
    df.append(pd.read_csv(f))
df = pd.concat(df).reset_index().drop("index",axis=1).drop("NONE",axis=1).rename({"_ASM_PERMUTATION":"N_FMA"},axis=1)
df["arch"] = "amd_zen3"
full_df = pd.concat([df,full_df])
df = []
for f in glob.glob("gold_*.csv"):  
    df.append(pd.read_csv(f))
df = pd.concat(df).reset_index().drop("index",axis=1).drop("NONE",axis=1).rename({"_ASM_PERMUTATION":"N_FMA"},axis=1)
df["arch"] = "intel_cascadelake_gold"
full_df = pd.concat([df,full_df])
full_df = full_df.reset_index().drop("index",axis=1)
full_df["vec_width"] = full_df["_ASM_VERSION"].apply(lambda x: int(x.split("_")[1]))
full_df["data_type"] = full_df["_ASM_VERSION"].apply(lambda x: x.split("_")[0])
full_df.to_csv("data_consolidated.csv")
#df = pd.read_csv("data_consolidated.csv")

full_df
```
